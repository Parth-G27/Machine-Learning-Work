## K Means Clustering Algorithms Implementation


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
```

```python
# Number of data points = 1000, number of clusters = 3
X,y = make_blobs(n_samples=1000, centers=3, n_features=2, random_state=22)
```

```python
X.shape
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.cluster import KMeans
```

```python
# Elbow method to select K 
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel("no. of clusters")
plt.ylabel("WCSS Value")
plt.xticks(range(1,11))
plt.show()
```

### From the diagram we can see the value of WCSS stabilizes at k = 3

```python
kmeans = KMeans(n_clusters=3,init='k-means++')
```

```python
y_labels = kmeans.fit_predict(X_train)
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=y_labels)
```

```python
y_test_label = kmeans.predict(X_test)
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_test_label)
```
